## Extracting Seasonal Crops

```python
kharif = [
    "rice", "maize", "cotton", "jute",
    "pigeonpeas", "mungbean", "blackgram", "mothbeans"
]

rabi = [
    "chickpea", "kidneybeans", "lentil"
]

zaid = [
    "watermelon", "muskmelon"
]

annual = [
    "banana", "mango", "grapes", "apple",
    "orange", "papaya", "coconut",
    "coffee", "pomegranate"
]
```

```python
kharif_crops = df[df["label"].isin(kharif)]
rabi_crops = df[df["label"].isin(rabi)]
zaid_crops = df[df["label"].isin(zaid)]
annual_crops = df[df["label"].isin(annual)]
```

```python
print("Kharif Crops:")
print(kharif_crops["label"].unique())

print("\nRabi Crops:")
print(rabi_crops["label"].unique())

print("\nZaid Crops:")
print(zaid_crops["label"].unique())

print("\nAnnual Crops:")
print(annual_crops["label"].unique())
```

```python
print("Number of Kharif samples :", len(kharif_crops))
print("Number of Rabi samples   :", len(rabi_crops))
print("Number of Zaid samples   :", len(zaid_crops))
print("Number of Annual samples :", len(annual_crops))
```

 # Handling Outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.boxplot(data=df[['N','P','K','temperature','humidity','ph','rainfall']])
plt.title("Box Plot for Outlier Detection")
plt.show()
```

```python
Q1 = df[['N','P','K','temperature','humidity','ph','rainfall']].quantile(0.25)
Q3 = df[['N','P','K','temperature','humidity','ph','rainfall']].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df[['N','P','K','temperature','humidity','ph','rainfall']] < (Q1 - 1.5 * IQR)) |
          (df[['N','P','K','temperature','humidity','ph','rainfall']] > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Dataset shape after outlier removal:", df.shape)
```
